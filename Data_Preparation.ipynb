```python
from nltk.tokenize import word_tokenize
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
sentence_1 = 'it was a good practice for us'
sentence_2 = 'it was also good to know about it'
sentence_3 = 'for all of us they have done a good practice'
```

```python
word = sentence_1.split() + sentence_2.split() + sentence_3.split()

set(word)
```

```python
words = word_tokenize(sentence_1.lower()) + word_tokenize(sentence_2.lower()) + word_tokenize(sentence_3.lower())
set(words)
```

```python
tokens = set(words)
```

```python
list(tokens)
```

```python
df = pd.DataFrame(columns=list(tokens), index=[1, 2, 3])
df
```

```python
token1 = word_tokenize(sentence_1.lower())
token2 = word_tokenize(sentence_2.lower())
token3 = word_tokenize(sentence_3.lower())
```

```python
counts1 = [token1.count(x) for x in df.columns]
counts2 = [token2.count(x) for x in df.columns]
counts3 = [token3.count(x) for x in df.columns]
```

```python
counts1, counts2, counts3
```

```python
df
```

```python
df.iloc[0,]=counts1
df.iloc[1,]=counts2
df.iloc[2,]=counts3
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer()
```

```python
mydata = cv.fit_transform([sentence_1, sentence_2, sentence_3])
```

```python
mydata.toarray()
```

```python
cv.get_feature_names_out()
```

```python
df = pd.DataFrame(data=mydata.toarray(),
                  columns=cv.get_feature_names_out())
```
